```python
import torch
x = torch.randn(32, 1, 64, 64)
# Flattening from the second dimension (index 1) onward
flattened = torch.flatten(x, start_dim=1)
print(f"Flattened shape: {flattened.shape}") 
print(f"Input features size: {flattened.shape[1]}")
```

```python
import torchaudio.transforms as T
import torch

waveform = torch.randn(1, 16000) # 1 second of audio
mel_spectrogram = T.MelSpectrogram(
    sample_rate=16000, n_fft=400, hop_length=160, n_mels=64
)(waveform)
print(f"Mel-Spectrogram shape: {mel_spectrogram.shape}")
```

```python
import torch.nn as nn
# Conv layer with p=1, k=3 keeps dimensions same: (64 + 2*1 - 3)/1 + 1 = 64
conv = nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1)
# MaxPool with k=2, s=2 halves dimensions: 64 / 2 = 32
pool = nn.MaxPool2d(kernel_size=2, stride=2)

x = torch.randn(1, 1, 64, 64)
output = pool(conv(x))
print(f"Output spatial shape: {output.shape[2:]}")
```

```python
linear = nn.Linear(128, 64)
# Parameters = (Weights: 128 * 64) + (Biases: 64)
params = sum(p.numel() for p in linear.parameters())
print(f"Total parameters: {params}")
```

```python
import math
total_files = 1050
batch_size = 32
# If drop_last=False, we round up to include the partial final batch
num_batches = math.ceil(total_files / batch_size)
print(f"Batches per epoch: {num_batches}")
```

```python
import torch.nn.functional as F
logits = torch.tensor([[2.5, 1.0, 0.1]])
target = torch.tensor([0])
loss = F.cross_entropy(logits, target)
print(f"Loss: {loss.item():.3f}")
```

```python
# New Weight = Old Weight - (Learning Rate * Gradient)
new_weight = 0.5 - (0.01 * 0.2)
print(f"New weight: {new_weight}")
```
